```python
df1=sns.countplot(x='top_track_popularity',data=df)

for bars in df1.containers:
    df1.bar_label(bars)
sns.set(rc={'figure.figsize':(25,10)})

plt.show()
```

Track no. 56 has highest count.


```python
df_1=df.head(25)
df_1
```

```python
df.groupby(['followers'], as_index=False)['top_track_popularity'].sum().sort_values(by='top_track_popularity',
                                                                   ascending=False)


df2=sns.lineplot(x="followers",y="top_track_popularity",data=df_1,palette="Accent")
for bars in df2.containers:
    df2.bar_label(bars)

sns.set(rc={'figure.figsize':(25,5)})
plt.show()
```

track.no. 80 has highest followers

```python
df3=sns.displot(df["top_track_duration_ms"])


plt.show()
```

```python
df4= sns.barplot(x='artist_name',y='popularity', data=df_1)

for bars in df4.containers:
    df4.bar_label(bars)
    

sns.set(rc={'figure.figsize':(25,5)})
plt.show()
```

popcan's popularity is higher 

```python
popcan's popularity is higher, followed by Queen Bee, Charly black and Freshlyground
```

```python
df5=sns.lineplot(x="artist_name",y="genres",data=df_1, size=10,palette="Accent",alpha=1)


for bars in df5.containers:
    df5.bar_label(bars)
    
sns.set(rc={'figure.figsize':(35,15)})
plt.legend()
plt.title("artist genres")
plt.show()
```

['anime rock', 'dance rock', 'j-rock'] by QUEEN BEE
